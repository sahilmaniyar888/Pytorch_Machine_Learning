```python
# Grouping data: A dictionary holding 3D coordinates
point_a = {"x": 1.5, "y": 2.0, "z": 0.5}
point_b = {"x": 4.0, "y": 2.0, "z": 3.0}

def calculate_spatial_distance(p1, p2):
    """Calculates Euclidean distance between two 3D points."""
    dx = p1["x"] - p2["x"]
    dy = p1["y"] - p2["y"]
    dz = p1["z"] - p2["z"]
    
    return (dx**2 + dy**2 + dz**2) ** 0.5

distance = calculate_spatial_distance(point_a, point_b)
print(f"The distance between Point A and Point B is: {distance:.2f}")
```

```python
# 1. We define the Blueprint using the 'class' keyword
class CADNode:
    
    # 2. The Constructor: This runs automatically when a node is created.
    # Notice 'self' is always the first parameter!
    def __init__(self, start_x, start_y, start_z):
        
        # We take the outside data (start_x) and lock it into the object's personal memory (self.x)
        self.x = start_x
        self.y = start_y
        self.z = start_z
        self.is_processed = False 
        
        print(f"New node created in memory at: ({self.x}, {self.y}, {self.z})")

    # 3. The Behavior (Method): A function built inside the object.
    def scale_node(self, scale_factor):
        
        # The object uses 'self' to look at its own internal memory, 
        # applies the math, and overwrites its own memory.
        self.x = self.x * scale_factor
        self.y = self.y * scale_factor
        self.z = self.z * scale_factor
        self.is_processed = True
        
        print(f"Node scaled! New coordinates: ({self.x}, {self.y}, {self.z})")

# --- END OF BLUEPRINT ---

# 4. Now we build actual Objects (Instances) from the blueprint
print("--- Creating Objects ---")
node_A = CADNode(start_x=1.0, start_y=2.0, start_z=0.5)
node_B = CADNode(start_x=10.0, start_y=10.0, start_z=5.0)

# 5. We trigger the built-in behaviors using the dot (.) operator
print("\n--- Triggering Behaviors ---")
# When node_A calls this, 'self' ensures only node_A's memory changes. node_B is safe.
node_A.scale_node(scale_factor=3.0)
```

```python
import torch
import torch.nn as nn

# Our custom AI model inherits from PyTorch's nn.Module
class SpatialFeatureExtractor(nn.Module):
    
    def __init__(self, input_size, hidden_size):
        # We MUST call super() so PyTorch can set up the invisible gradient tracking
        super().__init__() 
        
        # Now we define our specific network layers
        self.layer_1 = nn.Linear(input_size, hidden_size)
        self.activation = nn.ReLU()
        
    def forward(self, x):
        # We define how the data flows through our specific layers
        x = self.layer_1(x)
        return self.activation(x)

# Instantiating the model
# Because it inherited from nn.Module, it secretly has hundreds of hidden PyTorch methods.
ai_model = SpatialFeatureExtractor(input_size=128, hidden_size=64)

# For example, we can instantly move the entire model to a GPU using an inherited method!
# We didn't write .to(), we inherited it.
# ai_model.to("cuda")
print("Model instantiated with layers:", ai_model.layer_1)
print("Model is on device:", next(ai_model.parameters()).device)
```

```python
# 1. THE PARENT CLASS (The Foundation)
class BaseGeometry:
    def __init__(self, node_id):
        self.node_id = node_id
        self.is_rendered = False
        
    def render(self):
        self.is_rendered = True
        print(f"[{self.node_id}] Base geometry rendered to space.")

# 2. THE CHILD CLASS (The Specialized Version)
# By putting BaseGeometry in the parentheses, we inherit everything!
class ExtrudedCylinder(BaseGeometry):
    
    def __init__(self, node_id, radius, height):
        # 3. THE PORTAL: We trigger the Parent's setup first
        # This automatically creates self.node_id and self.is_rendered for us!
        super().__init__(node_id) 
        
        # 4. Now we add the specialized memory that only a cylinder needs
        self.radius = radius
        self.height = height
        
    def calculate_volume(self):
        # A brand new behavior just for this child
        pi = 3.14159
        return pi * (self.radius ** 2) * self.height

# --- Testing the Logic ---

print("--- Creating the Child Object ---")
# We only instantiate the Child, but watch what happens...
part_001 = ExtrudedCylinder(node_id="CYL-99", radius=5.0, height=10.0)

print("\n--- Triggering Inherited and Specialized Behaviors ---")
# We never wrote a render() function inside ExtrudedCylinder...
# But because of Inheritance, it has it anyway!
part_001.render() 

# And it has its own unique math too
vol = part_001.calculate_volume()
print(f"[{part_001.node_id}] Volume: {vol}")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# 1. THE CLASS & INHERITANCE
# We inherit from PyTorch's official 'Dataset' blueprint. 
# This tells PyTorch: "Treat my custom class exactly like your internal data engines."
class OrionFlowDataset(Dataset):
    
    # 2. THE CONSTRUCTOR (State)
    def __init__(self, text_prompts, point_clouds):
        super().__init__() # Initializes the PyTorch parent class
        
        # We store the massive lists of data inside the object's personal memory
        self.prompts = text_prompts
        self.geometries = point_clouds
        
    # 3. BUILT-IN PYTORCH BEHAVIOR 1
    # PyTorch requires this method so it knows exactly how much data exists.
    def __len__(self):
        return len(self.prompts)
        
    # 4. BUILT-IN PYTORCH BEHAVIOR 2
    # PyTorch calls this automatically when it needs exactly ONE specific 3D model.
    def __getitem__(self, index):
        # Fetch the specific text and 3D data based on the requested index
        prompt = self.prompts[index]
        raw_geometry = self.geometries[index]
        
        # Simulate preprocessing: Scaling the 3D point cloud by a factor of 2.0
        # Because raw_geometry is a PyTorch Tensor object, it knows how to multiply itself!
        processed_geometry = raw_geometry * 2.0 
        
        # We return a dictionary mapping the text to the CAD data
        return {"text": prompt, "cad_data": processed_geometry}


# --- THE EXECUTION (Running the Engine) ---

print("1. Simulating raw data loading...")
# Let's pretend these are our training files.
# 3 prompts, and 3 dummy 3D point clouds (each having 5 vertices with X, Y, Z coordinates).
dummy_prompts = ["a simple gear", "a connecting rod", "a mounting bracket"]
dummy_point_clouds = [torch.rand(5, 3), torch.rand(5, 3), torch.rand(5, 3)]

print("2. Instantiating the Custom Object...")
# We build the actual object from our blueprint
training_data_engine = OrionFlowDataset(dummy_prompts, dummy_point_clouds)

print("3. Wrapping it in a DataLoader for Batching...")
# DataLoader is another PyTorch class. It takes our custom dataset and handles
# the complex logic of shuffling it and grouping it into batches of 2.
dataloader = DataLoader(training_data_engine, batch_size=2, shuffle=True)

print("4. Starting the Training Loop...\n")
# The 'for' loop pulls data through the DataLoader. 
# Notice the accumulator variable (batch_num) tracking our progress!
batch_num = 1
for batch in dataloader:
    print(f"--- Processing Batch Number {batch_num} ---")
    print(f"Prompts in this batch: {batch['text']}")
    
    # .shape is a property of the Tensor object telling us its dimensions
    print(f"CAD Data dimensions [Batch Size, Vertices, Coordinates]: {batch['cad_data'].shape}\n")
    
    batch_num = batch_num + 1

print("Training pipeline complete!")
```

```python
def caluclate_bmi(weight: float, height: float) -> float:
    """Calculate Body Mass Index (BMI) given weight in kg and height in meters."""
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    bmi = weight / (height ** 2)
    return bmi  

weight = 70  # kg
height = 1.75  # meters
y = caluclate_bmi(weight, height)
print(f"Person has a BMI of {y:.2f}")
```

```python
import time

# 1. We build the Powder Coating Station (The Decorator)
def performance_tracker(func):
    
    # 2. We build the actual Wrapper Box
    # *args and **kwargs ensure this box can wrap ANY function perfectly.
    def wrapper_box(*args, **kwargs):
        
        # Start the stopwatch (The Pre-Coating)
        start_time = time.time()
        
        # Execute the original, pure mathematical function
        result = func(*args, **kwargs)
        
        # Stop the stopwatch (The Post-Coating)
        end_time = time.time()
        print(f"[DIAGNOSTIC] Function '{func.__name__}' took {end_time - start_time:.4f} seconds.")
        
        # Hand the final result back
        return result
        
    return wrapper_box

# --- Applying the Decorator ---

# We just drop the @ symbol on top. Python handles the rest.
@performance_tracker
def generate_heavy_cad_file(vertices):
    print("Generating geometry...")
    time.sleep(1.5) # Simulating a heavy 1.5-second math operation
    return f"Mesh with {vertices} vertices ready."

# When we trigger this, it invisibly routes through the 'wrapper_box' first!
final_file = generate_heavy_cad_file(10000)
```
