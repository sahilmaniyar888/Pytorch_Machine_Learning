```python
import numpy as np

A = np.array([ 
    [30,70,69], 
    [86,49,39], 
    [44,85,96], 
    [17,78,69],
    [56,89,77] 
], dtype=int)

# Correct for STUDENTS (Left to Right)
print("Student Averages:", np.mean(A, axis=1)) 
# Expected: [56.33, 58.0, 75.0, 54.66, 74.0]

# Correct for TOP SUBJECT per Student
print("Best Subject Index:", np.argmax(A, axis=1))
# Expected: [1, 0, 2, 1, 1] 
# (Student 1's best was subject 1, Student 2's best was subject 0, etc.)
```

```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.6, 0.5, 0.4, 0.35]
val_loss = [1.0, 0.7, 0.6, 0.5, 0.45]
```

```python
plt.plot(epochs, train_loss, label='Train Loss')

plt.plot(epochs, val_loss, label='Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.grid(True)
```
